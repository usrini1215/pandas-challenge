```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(file_to_load)
purchase_df
#purchase_df["SN"].head()
# command to export to output file
#file_one_df.to_csv("Output/fileOne.csv", index=False, header=True)
```

## Player Count

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
unique_items_df = purchase_df["Item ID"].value_counts()
average_price = round(purchase_df["Price"].mean(),2)
number_purchase = len(purchase_df)
total_revenue = purchase_df["Price"].sum()

summary_df = pd.DataFrame({"Number of Unique Items" : [len(unique_items_df)],
                           "Average Price" : [average_price],
                           "Number of Purchases" : [number_purchase],
                           "Total Revenue" : [total_revenue]
                          })

summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed



```python
#creating a unique list with Player and Gender and dropping the duplicate players.
unique_df = (purchase_df[["SN","Gender"]]).drop_duplicates()

#Gives the totals for Gendrer
percent_df = pd.DataFrame(unique_df["Gender"].value_counts())

#Total gender to calculate the percentage
total_gender = unique_df["Gender"].count()

#old logic, not using this!!!!
#male = unique_df["Gender"].value_counts()['Male']
#female = unique_df["Gender"].value_counts()['Female']
#non_gender_specific = total_gender - male - female
#male_percent = (male/total_gender) * 100
#female_percent = (female/total_gender) * 100
#non_gender_specific_percent = (non_gender_specific/total_gender) * 100

#display summary
percent_df['Percentage of Players'] = round((percent_df['Gender'] / total_gender) * 100,2)
percent_df.rename(columns ={'Gender':'Total Count'},inplace = True)
percent_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_grp_df1 = pd.DataFrame(purchase_df.groupby(["Gender"])[["Price"]].mean())
purchase_grp_df1.rename(columns ={'Price':'Average Purchase Price'},inplace = True)

purchase_grp_df2 = pd.DataFrame(purchase_df.groupby(["Gender"])[["Price"]].sum())
purchase_grp_df2.rename(columns ={'Price':'Total Purchase Value'},inplace = True)

purchase_grp_df3 = pd.DataFrame(purchase_df.groupby(["Gender"])[["Price"]].count())
purchase_grp_df3.rename(columns ={'Price':'Purchase Count'},inplace = True)

purchase_grp_df4 = pd.DataFrame(unique_df["Gender"].value_counts())


purchase_grp_df = [purchase_grp_df3, purchase_grp_df1, purchase_grp_df2,purchase_grp_df4]
result = pd.concat(purchase_grp_df, axis=1, sort=False)
result["Gender"] = round(result['Total Purchase Value']/result['Gender'],2)
result["Average Purchase Price"] = round(result['Average Purchase Price'],2)

result.rename(columns ={'Gender':'Avg Total Purchase per Person'},inplace = True)

result
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create the bins in which Data will be held 
bins = [0, 9, 14, 19, 24, 29, 34, 39,100]
# Create the names for bins
group_names = ["<10","10-14", "15-19", "20-24", "25-29", "30-34", "35-39","40+"]

unique_age_df = (purchase_df[["SN","Age"]]).drop_duplicates()
unique_age_df["Age Ranges"] = pd.cut(unique_age_df["Age"], bins, labels=group_names, include_lowest=True)

unique_age_df = unique_age_df.groupby("Age Ranges").count()
unique_age_df["Percentage of Players"] = round(unique_age_df["Age"]/576*100,2)

unique_age_df.rename(columns ={'SN':'Total Count'},inplace = True)
del unique_age_df["Age"]

unique_age_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Create the bins in which Data will be held 
bins = [0, 9, 14, 19, 24, 29, 34, 39,100]
# Create the names for bins
group_names = ["<10","10-14", "15-19", "20-24", "25-29", "30-34", "35-39","40+"]

#new_age_df = (purchase_df[["SN","Age"]]
purchase_age_df = purchase_df
purchase_age_df["Age Ranges"] = pd.cut(purchase_df["Age"], bins, labels=group_names, include_lowest=True)

purchase_age_df1 = pd.DataFrame(purchase_age_df.groupby(["Age Ranges"])[["Price"]].mean())
purchase_age_df1.rename(columns ={'Price':'Average Purchase Price'},inplace = True)

purchase_age_df2 = pd.DataFrame(purchase_age_df.groupby(["Age Ranges"])[["Price"]].sum())
purchase_age_df2.rename(columns ={'Price':'Total Purchase Value'},inplace = True)

purchase_age_df3 = pd.DataFrame(purchase_age_df.groupby(["Age Ranges"])[["Price"]].count())
purchase_age_df3.rename(columns ={'Price':'Purchase Count'},inplace = True)

unique_age_range_df = (purchase_df[["SN","Age"]]).drop_duplicates()
unique_age_range_df["Age Ranges"] = pd.cut(purchase_df["Age"], bins, labels=group_names, include_lowest=True)

purchase_age_df4 = pd.DataFrame(unique_age_range_df["Age Ranges"].value_counts())


purchase_merge_df = [purchase_age_df3, purchase_age_df1, purchase_age_df2,purchase_age_df4]
result2 = pd.concat(purchase_merge_df, axis=1, sort=False)
result2["Age Ranges"] = round(result2['Total Purchase Value']/result2['Age Ranges'],2)
result2["Average Purchase Price"] = round(result2['Average Purchase Price'],2)

result2.rename(columns ={'Age Ranges':'Avg Total Purchase per Person'},inplace = True)

result2

#purchase_age_df.set_index("Age Ranges")

#purchase_age_df = purchase_age_df.drop(['Purchase ID', 'SN', 'Age', 'Gender', 'Item ID', 'Item Name', 'Price'],axis = 1)

#purchase_age_df.rename(columns ={'SN':'Total Count'},inplace = True)
#del purchase_age_df["Age"]
```
