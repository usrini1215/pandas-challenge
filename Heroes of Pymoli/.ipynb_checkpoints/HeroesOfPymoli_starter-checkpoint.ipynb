```python
# Create the bins in which Data will be held 
bins = [0, 9, 14, 19, 24, 29, 34, 39,100]
# Create the names for bins
group_names = ["<10","10-14", "15-19", "20-24", "25-29", "30-34", "35-39","40+"]

#unique df on player and age and add Age ranges to bin
unique_age_df = (purchase_df[["SN","Age"]]).drop_duplicates()
unique_age_df["Age Ranges"] = pd.cut(unique_age_df["Age"], bins, labels=group_names, include_lowest=True)

#group it by age ranges and get count
unique_age_df = unique_age_df.groupby("Age Ranges").count()
unique_age_df["Percentage of Players"] = round(unique_age_df["Age"]/576*100,2).map("{0:.2f}%".format)


unique_age_df.rename(columns ={'SN':'Total Count'},inplace = True)
del unique_age_df["Age"]

unique_age_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Retrieve the Item ID, Item Name, and Item Price columns
items_price_df = purchase_df.groupby(["Item ID", "Item Name"])[["Price"]].mean()

#grouping by item id and name; and get the total purchase count
items_df = purchase_df.groupby(["Item ID", "Item Name"], as_index=False)[["Purchase ID"]].count()
items_df = items_df.rename(columns = {"Purchase ID": "Purchase Count"})
items_df = items_df.set_index(["Item ID","Item Name"])

#Merge the above 2 DF's to get the average price for the total purchase
merged_df = [items_df, items_price_df]
result_df = pd.concat(merged_df, axis=1, sort=False)
result_df = result_df.sort_values("Purchase Count", ascending=False)
result_df["Total Purchase Value"] = result_df["Purchase Count"] * result_df ["Price"]
result_df["Price"] = result_df["Price"].map("${:.2f}".format)
result_df["Total Purchase Value"] = result_df["Total Purchase Value"].map("${:.2f}".format)


result_df.head()

# price = items_price_df.loc[items_price_df['Item ID'] == 92, "Price"]
# type(price)
# sum(price)

# tst = purchase_df[["Item Name","Purchase ID","Price"]]
# tst.set_index("Item Name")
#price = tst.loc[tst['Item Name'] == "Oathbreaker, Last Hope of the Breaking Storm", "Price"]
# price
```
